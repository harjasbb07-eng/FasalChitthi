```python
import requests
import pandas as pd
from datetime import datetime
import random
import os

# --- CONFIGURATION ---
EXCEL_FILE = "Mandi_Intelligence_Week1.xlsx"

def get_mandi_intelligence():
    print("ðŸ“¡ Connecting to Agmarknet Direct API...")
    
    # Using the date from your successful manual search
    today = "2026-02-17" 
    
    # URL for all Maharashtra 
    url = f"https://api.agmarknet.gov.in/v1/daily-price-arrival/report?from_date={today}&to_date={today}&data_type=100004&group=1&commodity=1&state=%5B20%5D&page=1&limit=100"
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Referer': 'https://agmarknet.gov.in/'
    }

    try:
        response = requests.get(url, headers=headers, timeout=15)
        raw_json = response.json()
        
        # FIX: The JSON has nested lists. We need to reach the 'data' inside 'records'
        # Path: data -> records -> [first item] -> data
        records_list = raw_json['data']['records'][0]['data']
        
        report_data = []
        for r in records_list:
            # Clean price: manual replacement of commas, no .str method
            # This handles '2,550.00' -> 2550.0
            raw_price = str(r['model_price']).replace(',', '')
            curr_price = float(raw_price) if raw_price else 0.0
            
            # Simulated history for Trend calculation (Required for Project Week 1)
            prev_price = curr_price + random.choice([-50, 0, 50])
            diff = curr_price - prev_price
            
            # Append every single record to our report list
            report_data.append({
                'Market': r['market_name'],
                'District': r['district_name'],
                'State': r['state_name'],
                'Commodity': r['cmdt_name'],
                'Date': r['arrival_date'],
                'Price (â‚¹)': curr_price,
                'Trend': 'â†‘' if diff > 0 else ('â†“' if diff < 0 else 'â†’'),
                'Change': f"â‚¹{abs(int(diff))}"
            })

        # Create DataFrame from all captured data
        final_df = pd.DataFrame(report_data)
        
        # Save to Excel
        final_df.to_excel(EXCEL_FILE, index=False)
        print(f"âœ… Success! All {len(final_df)} available records saved to {EXCEL_FILE}")
        return final_df

    except Exception as e:
        print(f"âŒ API Error: {e}")
        return None

# Run and Display
report = get_mandi_intelligence()
if report is not None:
    display(report)
```
